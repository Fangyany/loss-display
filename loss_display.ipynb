```python
loss_g_max = []
loss_g_min = []
loss_g_mid = []

for i in range(25):
    list = [loss_only_generator2[i], loss_only_generator3[i], loss_only_generator4[i], loss_only_generator5[i], loss_only_generator6[i]]
    list.sort()
    max = list[-1]
    min = list[0]
    mid = (list[-1]+list[0])/2
    loss_g_max.append(max)
    loss_g_min.append(min)
    loss_g_mid.append(mid)
    
    
loss_gan_max = []
loss_gan_min = []
loss_gan_mid = []

for i in range(25):
    list = [ loss_GAN2[i], loss_GAN3[i], loss_GAN4[i], loss_GAN5[i]]
    list.sort()
    max = list[-1]
    min = list[0]
    mid = (list[-1]+list[0])/2
    loss_gan_max.append(max)
    loss_gan_min.append(min)
    loss_gan_mid.append(mid)
```

```python

import seaborn as sns

fig, ax = plt.subplots()
a = range(len(loss_g_mid))
sns.set()

plt.ylabel('average displacement error')
# plt.ylabel('final displacement error')

plt.xlabel('iteration number')


ax.plot(loss_g_mid,'--', color="#c72e29", lw=2, label='baseline')
ax.fill_between(a, loss_g_max, loss_g_min,  color="tomato", alpha = 0.5)

# fig, ax = plt.subplots()
ax.plot(loss_gan_mid, color = 'darkgreen', lw=2, label='GAN-based')
ax.fill_between(a, loss_gan_max, loss_gan_min, color = 'mediumseagreen', alpha = 0.5)

plt.legend()
plt.rcParams['figure.figsize'] = (6,6)

plt.savefig('ADE.eps')
```
